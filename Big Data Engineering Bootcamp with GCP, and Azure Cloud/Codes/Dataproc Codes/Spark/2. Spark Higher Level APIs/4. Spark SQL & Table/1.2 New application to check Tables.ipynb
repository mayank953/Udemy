```python
from pyspark.sql import SparkSession



spark = SparkSession.builder \
.appName('Table Verification') \
.enableHiveSupport() \
.getOrCreate()
```

```python
spark_new_app_session = spark.newSession()
```

```python
spark_new_app_session.sql('show tables').show()
```

```python
spark_new_app_session.sql('show tables in global_temp').show()
```

```python
spark_new_app_session.sql('select * from global_temp.global_customers limit 5').show()
```
